<h2 align='center' style='color:blue'>Dropout Regularization In Deep Neural Network</h2>

This is a dataset that describes sonar chirp returns bouncing off different services. The 60 input variables are the strength of the returns at different angles. It is a **binary classification problem** that requires a model to differentiate rocks from metal cylinders.

Dataset information: https://archive.ics.uci.edu/ml/datasets/Connectionist+Bench+(Sonar,+Mines+vs.+Rocks)
Download it from here: https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./sonar_dataset.csv", header=None)
df.sample(5)
```

```python
df.shape
```

```python
# check for nan values
df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts() # label is not skewed
```

```python
X = df.drop(60, axis=1)
y = df[60]
y.head()
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # R --> 1 and M --> 0
```

```python
y.value_counts()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.head()
```

# Using Deep Learning Model

## Model without Dropout Layer

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

Training Accuracy >>> Test Accuracy

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

### Model with Dropout Layer

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test, y_test)
```

Training Accuracy is still good but Test Accuracy Improved

```python
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

**You can see that by using dropout layer test accuracy increased from 0.77 to 0.81**

```python

from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
```

```python

model_bn = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    BatchNormalization(),
    keras.layers.Dense(30, activation='relu'),
    BatchNormalization(),
    keras.layers.Dense(15, activation='relu'),
    BatchNormalization(),
    keras.layers.Dense(1, activation='sigmoid')
])

model_bn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Define Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Include validation split for early stopping monitoring
history_bn = model_bn.fit(X_train, y_train, epochs=100, batch_size=8, validation_split=0.2, callbacks=[early_stopping])
```

```python

evaluation_bn = model_bn.evaluate(X_test, y_test)
```

```python

def plot_learning_curve(history, title='Learning Curve'):
    epochs = range(1, len(history.history['accuracy']) + 1)
    plt.figure(figsize=(10, 5))
    plt.plot(epochs, history.history['loss'], 'b-', label='Training Loss')
    plt.plot(epochs, history.history['val_loss'], 'r--', label='Validation Loss')
    plt.title(title)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.plot(epochs, history.history['accuracy'], 'b-', label='Training Accuracy')
    plt.plot(epochs, history.history['val_accuracy'], 'r--', label='Validation Accuracy')
    plt.title(title)
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

# Plot learning curves for each model
plot_learning_curve(history, title='Dropout Model Learning Curve')
plot_learning_curve(history_bn, title='Batch Norm + Early Stopping Model Learning Curve')
```
